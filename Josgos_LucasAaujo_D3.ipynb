```python
import pandas as pd

# Dados dos Jogos
nome_jogo = ["Little nightmares", "Valorant", "LOL", "R6", "minecraft", "GTAV"]
criadores = ["Konami", "RIOT games", "ROT games", "Ubsoft", "Mojang", "Rockstar games"]
generos = ["Suspence", "FPS", "Estrategia", "FPS", "Sandbox", "Sandbox"]
ano_lancamento = [2017, 2020, 2013, 2014, 2011, 2013]
duracao_jogo = [4.00, 1.0, 1.0, 0.30, 18.00, 20.00]
porte = ["pc/ps4/xbox","Pc","Pc","Pc/ps4/xbox","Pc/ps4/xbox","Pc/ps4/xbox"]

# Crie um dicionário contendo dados dos jogos
dados_jogos = {
    'nome do Jogo': nome_jogo,
    'Criador': criadores,
    'genero': generos,
    'Ano de lançamento': ano_lancamento,
    'Duração (Horas)': duracao_jogo,
    'Porte do game' : porte
}

# Crie um DataFrame a partir do dicionário
jogos_df = pd.DataFrame(dados_jogos)
print(jogos_df)

print()

# Exibir o DataFrame
display(jogos_df)

#--2
print(f"Tamanho do DataFrame: {jogos_df.shape[0]} linhas x {jogos_df.shape[1]} colunas")

#--3
x = 2
item = jogos_df.iloc[x]
print(f"\nCaracterísticas do item na linha {x}:")
print(item)

#--4
if jogos_df.empty:
    print("O DataFrame está vazio.")
else:
    print("O DataFrame não está vazio.")

#--5
print("\nOs 5 primeiros registros do DataFrame:")
print(jogos_df.head())

#--6
linha_para_excluir = 2
jogos_df = jogos_df.drop(linha_para_excluir).reset_index(drop=True)

#--7
novo_jogo = {
    'nome do Jogo': "Cyberpunk 2077",
    'Criador': "CD Projekt",
    'genero': "RPG",
    'Ano de lançamento': 2020,
    'Duração (Horas)': 40.00
}

jogos_df.loc[jogos_df.shape[0]] = novo_jogo
print("\nDataFrame após a adição de uma nova linha:")
print(jogos_df)

print(f"\nTamanho do DataFrame após a adição: {jogos_df.shape[0]} linhas x {jogos_df.shape[1]} colunas")

#--8
jogos_df_transposto = jogos_df.transpose()

print("\nDataFrame transposto:")
print(jogos_df_transposto)
print(f"\nTamanho do DataFrame transposto: {jogos_df_transposto.shape[0]} linhas x {jogos_df_transposto.shape[1]} colunas")

#--9
print("\nSomente a 1ª e a 2ª coluna do DataFrame:")
print(jogos_df[['nome do Jogo', 'Criador']])
```
